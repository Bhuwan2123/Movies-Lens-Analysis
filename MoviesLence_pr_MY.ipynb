# Stats Model

```python
import statsmodels.formula.api as sm
md = sm.ols(formula = 'Rating~Gender+Action+Adventure+Documentary+Age+Animation+Comedy+Crime+Horror+Horror+Romance+Thriller+War+Western',data = dt_2).fit()
md.summary()
```

Stats models shows that Age and Gender columns are important becaise there p values are greater less that 0.05

Splitting data in train and test

```python
x1 = dt_2.drop('Rating',axis = 1)
y1 = dt_2['Rating']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x1,y1,random_state = 1, train_size = 0.3)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# 4.	Develop an appropriate model to predict the movie ratings

# Model Building

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
```

## Predicting Values and evaluating them

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score as r2
print(mae(y_test,pred),mse(y_test,pred),r2(y_test,pred))
```
